# Selenium을 활용하여 날씨 데이터 크롤링

```python
from selenium import webdriver
from selenium.webdriver.support.select import Select
import time 
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.get("https://data.kma.go.kr/climate/StatisticsDivision/selectStatisticsDivision.do?pgmNo=158")

region = """//*[@id="btnStn"]"""
driver.find_element_by_xpath(region).click()

time.sleep(2)
choose_all = """//*[@id="ztree_1_check"]"""
driver.find_element_by_id("ztree_1_check").click()

finish = """//*[@id="sidetreecontrol"]/a"""
driver.find_element_by_xpath(finish).click()

# 일별 선택 
day = """//*[@id="dataFormCd"]/option[1]"""
driver.find_element_by_xpath(day).click()

# 시작하는 년도: 끝나는 년도와 동일
startYear= Select(driver.find_element_by_xpath('//*[@id="startYear"]'))
startYear.select_by_value('2021')

# 끝나는 년도: 시작하는 년도랑 동일 
endYear= Select(driver.find_element_by_xpath('//*[@id="endYear"]'))
endYear.select_by_value('2021')

# 시작하는 월: 끝나는 월과 동일 
startMonth= Select(driver.find_element_by_xpath('//*[@id="startMonth"]'))
startMonth.select_by_value('05')
    
# 끝나는 월: 시작하는 월과 동일 
endMonth= Select(driver.find_element_by_xpath('//*[@id="endMonth"]'))
endMonth.select_by_value('05')

# 시작하는 날: 끝나는 날과 동일 
startDate= Select(driver.find_element_by_xpath('//*[@id="startDay"]'))
startDate.select_by_value('05')

# 끝나는 날: 시작하는 날과 동일 = 하루 
endDate= Select(driver.find_element_by_xpath('//*[@id="endDay"]'))
endDate.select_by_value('05')

search = """//*[@id="schForm"]/div[3]/button"""
driver.find_element_by_xpath(search).click()

weather_df = pd.DataFrame([])
soup = BeautifulSoup(driver.page_source, "lxml")
table = soup.find_all('table')[1]
df = pd.read_html(str(table),header=0)
weather_df.append(df)
```

```python
def yearly_data(year): 
    
    driver = webdriver.Chrome('./chromedriver')
    driver.get("https://data.kma.go.kr/climate/StatisticsDivision/selectStatisticsDivision.do?pgmNo=158")

    region = """//*[@id="btnStn"]"""
    driver.find_element_by_xpath(region).click()

    time.sleep(2)
    choose_all = """//*[@id="ztree_1_check"]"""
    driver.find_element_by_id("ztree_1_check").click()

    finish = """//*[@id="sidetreecontrol"]/a"""
    driver.find_element_by_xpath(finish).click()

    # 일별 선택 
    day = """//*[@id="dataFormCd"]/option[1]"""
    driver.find_element_by_xpath(day).click()

    # 시작하는 년도: 끝나는 년도와 동일
    startYear = Select(driver.find_element_by_xpath('//*[@id="startYear"]'))
    startYear.select_by_value(f'{year}')

    # 끝나는 년도: 시작하는 년도랑 동일 
    endYear = Select(driver.find_element_by_xpath('//*[@id="endYear"]'))
    endYear.select_by_value(f'{year}')
    
    df = pd.DataFrame()

         
    for month in range(1,13): 
        time.sleep(0.2)
        startMonth = """//*[@id="startMonth"]/option[{}]"""
        driver.find_element_by_xpath(startMonth.format(month)).click()

        time.sleep(0.2)
        endMonth = """//*[@id="endMonth"]/option[{}]"""
        driver.find_element_by_xpath(endMonth.format(month)).click()

        for date in range(1,32):
            try:
                time.sleep(0.2)
                startDate= """//*[@id="startDay"]/option[{}]"""
                driver.find_element_by_xpath(startDate.format(date)).click()

                time.sleep(0.2)
                endDate= """//*[@id="endDay"]/option[{}]"""
                driver.find_element_by_xpath(endDate.format(date)).click()

                time.sleep(0.2)
                search = """//*[@id="schForm"]/div[3]/button"""
                driver.find_element_by_xpath(search).click()

                soup = BeautifulSoup(driver.page_source, "lxml")
                print(soup.find_all('table'))
                table = soup.find_all('table')[1]
                df1 = pd.read_html(str(table),header=0)
                df = df.append(df1)
                #df.columns = ["date", "avg_temp", "max_temp", "min_temp", "rain_fall"]

                if date is None: 
                    print("day is none")

            except: 
                df.append(np.nan)
                continue

    return df
    driver.quit()
```

```python
year_13 = yearly_data(2013)
year_13 = year_13.dropna(thresh=2) 
year_13.to_excel(r'2013weather.xlsx')
```
